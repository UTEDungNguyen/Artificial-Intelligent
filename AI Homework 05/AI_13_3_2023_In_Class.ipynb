```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
url = 'mtcars.csv'
data = pd.read_csv(url)
data.hist()
plt.show()
```

```python
correlation = data.corr()
print(correlation)
```

```python
sns.heatmap(correlation, annot = True)
```

```python
data.plot(kind = 'scatter', x = 'wt', y = 'mpg', figsize=(9,9), color = 'red')
```

```python
from sklearn import linear_model
```

```python
model = linear_model.LinearRegression()
model.fit(X = pd.DataFrame(data['wt']), y = data['mpg']) #qua trinh hoc va dieu chinh cho may
print(model.intercept_)  # hang so b
print(model.coef_) # he so goc a
model.score(X = pd.DataFrame(data['wt']), y = data['mpg']) #độ chính xác bao nhiêu %
```

```python
data.plot(kind = 'scatter', x = 'wt', y = 'mpg', figsize = (9,9), color='black', xlim = (0,7))
y_prid = model.predict(X = pd.DataFrame(data['wt']))
plt.plot(data['wt'], y_prid, color = 'blue')
```

```python
model = linear_model.LinearRegression() # Tạo mô hình mới
X = pd.DataFrame([data['wt'], data['wt']**2]).T
model.fit(X, y=data['mpg'])
print('Intercept của mô hình :', model.intercept_)
print('Coefficent của mô hình :', model.coef_)
print('Độ chính xác của mô hình :', model.score(X, y=data['mpg']))
```

```python
import numpy as np
range = np.arange(1.5, 5.5, 0.1)
data.plot(kind = 'scatter', x = 'wt', y = 'mpg', figsize = (9,9), color='black', xlim = (0,7))
predictor = pd.DataFrame([range, range**2]).T
y_prid = model.predict(X = predictor)
plt.plot(range, y_prid, color = 'blue')
```

```python
model = linear_model.LinearRegression() # Tạo mô hình mới
X = pd.DataFrame([data['wt'], data['wt']**2, data['wt']**3]).T
model.fit(X, y=data['mpg'])
print('Intercept của mô hình :', model.intercept_)
print('Coefficent của mô hình :', model.coef_)
print('Độ chính xác của mô hình :', model.score(X, y=data['mpg']))
```

**Advertising**

```python
import seaborn as sns
url = 'Advertising.csv'
data = pd.read_csv(url)
data.hist()
plt.show()
sns.pairplot(data) # Hiển thị data
```

```python
correlation = data.corr()
print(correlation)
```

```python
sns.heatmap(correlation, annot = True)
```

```python
data.plot(kind = 'scatter', x = 'TV', y = 'sales', figsize=(9,9), color = 'red')
```

```python
model = linear_model.LinearRegression()
model.fit(X = pd.DataFrame(data['TV']), y = data['sales']) #qua trinh hoc va dieu chinh cho may
print(model.intercept_)  # hang so b
print(model.coef_) # he so goc a
model.score(X = pd.DataFrame(data['TV']), y = data['sales']) #độ chính xác bao nhiêu %
```

```python
model = linear_model.LinearRegression() # Tạo mô hình mới
X = pd.DataFrame([data['TV'], data['TV']**2, data['TV']**3, data['TV']**4, data['TV']**5]).T
model.fit(X, y=data['sales'])
print('Intercept của mô hình :', model.intercept_)
print('Coefficent của mô hình :', model.coef_)
print('Độ chính xác của mô hình :', model.score(X, y=data['sales']))

import numpy as np
range = np.arange(data['TV'].min(), data['TV'].max(), 0.1)
data.plot(kind = 'scatter', x = 'TV', y = 'sales', figsize = (9,9), color='black', xlim = (data['TV'].min(),data['TV'].max()))
predictor = pd.DataFrame([range, range**2, range**3, range**4, range**5]).T
y_prid = model.predict(X = predictor)
plt.plot(range, y_prid, color = 'blue')
```
