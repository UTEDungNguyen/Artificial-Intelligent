```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import linear_model
import numpy as np
from sklearn.linear_model import LinearRegression
```

**Passenger**

```python
url = 'passenger.csv'
data = pd.read_csv(url)
data.hist()
plt.show()
sns.pairplot(data) # Hiển thị data
```

```python
correlation = data.corr()
print(correlation)
```

```python
sns.heatmap(correlation, annot = True)
```

```python
data.plot(kind = 'scatter', x = 'Miles(X)', y = 'Dollars (Y)', figsize=(9,9), color = 'red')
```

```python
model = linear_model.LinearRegression()
model.fit(X = pd.DataFrame(data['Miles(X)']), y = data['Dollars (Y)']) #qua trinh hoc va dieu chinh cho may
print(model.intercept_)  # hang so b
print(model.coef_) # he so goc a
model.score(X = pd.DataFrame(data['Miles(X)']), y = data['Dollars (Y)']) #độ chính xác bao nhiêu %
```

```python
model = linear_model.LinearRegression() # Tạo mô hình mới
X = pd.DataFrame([data['Miles(X)'], data['Miles(X)']**2, data['Miles(X)']**3]).T
model.fit(X, y=data['Dollars (Y)'])
print('Intercept của mô hình :', model.intercept_)
print('Coefficent của mô hình :', model.coef_)
print('Độ chính xác của mô hình :', model.score(X, y=data['Dollars (Y)']))

import numpy as np
range = np.arange(data['Miles(X)'].min(), data['Miles(X)'].max(), 0.1)
data.plot(kind = 'scatter', x = 'Miles(X)', y = 'Dollars (Y)', figsize = (9,9), color='black', xlim = (data['Miles(X)'].min(),data['Miles(X)'].max()))
predictor = pd.DataFrame([range, range**2, range**3]).T
y_prid = model.predict(X = predictor)
plt.plot(range, y_prid, color = 'blue')
```

**BodyFat**

```python
url = 'bodyfat.csv'
data = pd.read_csv(url)
data.hist()
plt.show()
sns.pairplot(data) # Hiển thị data
```

```python
correlation = data.corr()
print(correlation)
```

```python
sns.heatmap(correlation, linewidth=.5, annot = True)
```

```python
data.plot(kind = 'scatter', x = 'BodyFat', y = 'Density', figsize=(9,9), color = 'red')
```

```python
model = linear_model.LinearRegression()
model.fit(X = pd.DataFrame(data['BodyFat']), y = data['Density']) #qua trinh hoc va dieu chinh cho may
print(model.intercept_)  # hang so b
print(model.coef_) # he so goc a
model.score(X = pd.DataFrame(data['BodyFat']), y = data['Density']) #độ chính xác bao nhiêu %
```

```python
model = linear_model.LinearRegression() # Tạo mô hình mới
X = pd.DataFrame([data['BodyFat'], data['BodyFat']**2, data['BodyFat']**3, data['BodyFat']**4, data['BodyFat']**5]).T
model.fit(X, y=data['Density'])
print('Intercept của mô hình :', model.intercept_)
print('Coefficent của mô hình :', model.coef_)
print('Độ chính xác của mô hình :', model.score(X, y=data['Density']))

import numpy as np
range = np.arange(data['BodyFat'].min(), data['BodyFat'].max(), 0.1)
data.plot(kind = 'scatter', x = 'BodyFat', y = 'Density', figsize = (9,9), color='black', xlim = (data['BodyFat'].min(),data['BodyFat'].max()))
predictor = pd.DataFrame([range, range**2, range**3, range**4, range**5]).T
y_prid = model.predict(X = predictor)
plt.plot(range, y_prid, color = 'blue')
```
