```python
import pandas as pd
url = 'http://winterolympicsmedals.com/medals.csv'
dataframe = pd.read_csv(url)
print(dataframe)
```

**Chọn quốc gia để báo cáo bài tập là nước Đức (GER)**

**Số huy chương vàng của Đức (GER) qua các năm :**

```python
dataframe[(dataframe['NOC'] == 'GER')&(dataframe['Medal'] == 'Gold')].head(1000)
```

```python
fram = dataframe[(dataframe['NOC'] == 'GER')&(dataframe['Medal'] == 'Gold')].count()
print("Số huy chương vàng đạt được của Đức qua các năm:",fram['Medal'])
```

**Số huy chương bạc của Đức(GER) qua các năm :**

```python
dataframe[(dataframe['NOC'] == 'GER')&(dataframe['Medal'] == 'Silver')].head(1000)
```

```python
fram = dataframe[(dataframe['NOC'] == 'GER')&(dataframe['Medal'] == 'Silver')].count()
print("Số huy chương bạc đạt được của Đức qua các năm:",fram['Medal'])
```

**Số huy chương đồng của Đức (GER) qua các năm :**

```python
dataframe[(dataframe['NOC'] == 'GER')&(dataframe['Medal'] == 'Bronze')].head(1000)
```

```python
frame = dataframe[(dataframe['NOC'] == 'GER')&(dataframe['Medal'] == 'Bronze')].count()
print("Số huy chương đồng đạt được của Đức qua các năm:",frame['Medal'])
```

**Tổng số huy chương theo từng bộ môn của GER qua các năm:**

```python
Sport = dataframe[dataframe['NOC']=='GER']
Sport.groupby('Sport')['Medal'].count()
```

```python
Sport = dataframe.loc[dataframe.NOC == 'GER']
total = Sport.Sport.value_counts().count()
print('Tổng số bộ môn mà GER tham gia qua tất cả các năm : ',total)
```

**Các năm Đức tham dự Olympics mùa đông và có huy chương:**

```python
frame = dataframe.loc[dataframe.NOC == 'GER']
numberofyear = frame['Year'].unique()
print('Các năm tham dự Olympics mùa đông của Đức : ',numberofyear)
total = frame.Year.value_counts().count()
print('\n Tổng số năm tham dự Olympics của Đức : ',total)
```
